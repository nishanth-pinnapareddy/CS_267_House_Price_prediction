## Calculate Percentage of missing values for each varaiable in train data set

```python
for varaible in house_training_data.column_names():
    percent_missing_values = (float(house_training_data[varaible].num_missing())/ len(house_training_data[varaible])) * 100.0
    print ("{0}: \t{1}".format(varaible, percent_missing_values))
```

## Calculate the percentage of missing values in train data set

```python
num_of_missing_values = 0.0
total_no_of_values = 0.0
for varaible in house_training_data.column_names():
    num_of_missing_values += house_training_data[varaible].num_missing()
    total_no_of_values += len(house_training_data[varaible])
percent_missing_values = num_of_missing_values / total_no_of_values * 100.0
print "Percentage of Missing Values in Train Dataset: \t {0}".format(percent_missing_values)
```

## Calculate percentage of missing values for each variable in test data set

```python
for varaible in house_test_data.column_names():
    percent_missing_values = (float(house_test_data[varaible].num_missing())/ len(house_test_data[varaible])) * 100.0
    print ("{0}: \t{1}".format(varaible, percent_missing_values))
```

## Calculate percentage of missing values in test data set

```python
num_of_missing_values = 0.0
total_no_of_values = 0.0
for varaible in house_test_data.column_names():
    num_of_missing_values += house_test_data[varaible].num_missing()
    total_no_of_values += len(house_test_data[varaible])
percent_missing_values = num_of_missing_values / total_no_of_values * 100.0
print "Percentage of Missing Values in Test Dataset: \t {0}".format(percent_missing_values)
```

## Get Pandas data frame from SFrame

```python
pd_house_train_data = house_training_data.to_dataframe()
```

## Corelation of Numerical Variable with SalePrice

```python
corr = pd_house_train_data.select_dtypes(include = ['float64', 'int64']).corr()
```

```python
cor_dict = corr['SalePrice'].to_dict()
del cor_dict['SalePrice']
for sorted_tuple in sorted(cor_dict.items(), key = lambda item: -abs(item[1])):
    print("{0}: \t{1}".format(*sorted_tuple))
```

The housing price correlates strongly with OverallQual, GrLivArea(GarageCars), GargeArea, TotalBsmtSF, 1stFlrSF, FullBath, TotRmsAbvGrd, YearBuilt, YearRemodAdd, GargeYrBlt, MasVnrArea and Fireplaces. But some of those features are highly correlated among each others.

## Import seaborn and matplotlib for visualizing data

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

## Scatter plots between features having high corelation with Saleprice

```python
plt.figure(1)
f, axarr = plt.subplots(4, 2, figsize=(10, 9))
price = pd_house_train_data.SalePrice.values
axarr[0, 0].scatter(pd_house_train_data.OverallQual.values, price)
axarr[0, 0].set_title('OverallQual')
axarr[0, 1].scatter(pd_house_train_data.GrLivArea.values, price)
axarr[0, 1].set_title('GrLiveArea')
axarr[1, 0].scatter(pd_house_train_data.GarageArea.values, price)
axarr[1, 0].set_title('GarageArea')
axarr[1, 1].scatter(pd_house_train_data.TotalBsmtSF.values, price)
axarr[1, 1].set_title('TotalBsmtSF')
axarr[2, 0].scatter(pd_house_train_data['1stFlrSF'].values, price)
axarr[2, 0].set_title('1stFlrSF')
axarr[2, 1].scatter(pd_house_train_data.TotRmsAbvGrd.values, price)
axarr[2, 1].set_title('TotRmsAbvGrd')
axarr[3, 0].scatter(pd_house_train_data.MasVnrArea.values, price)
axarr[3, 0].set_title('MasVnrArea')
axarr[3, 1].scatter(pd_house_train_data.Fireplaces.values, price)
axarr[3, 1].set_title('Fireplaces')
f.text(-0.01, 0.5, 'Sale Price', va='center', rotation='vertical', fontsize = 12)
plt.tight_layout()
plt.show()
```

## Scatter plots between YearBuilt, YearReModAdd vs SalePrice

```python
fig = plt.figure(2, figsize=(9, 7))
plt.subplot(211)
plt.scatter(pd_house_train_data.YearBuilt.values, price)
plt.title('YearBuilt')

plt.subplot(212)
plt.scatter(pd_house_train_data.YearRemodAdd.values, price)
plt.title('YearRemodAdd')

fig.text(-0.01, 0.5, 'Sale Price', va = 'center', rotation = 'vertical', fontsize = 12)

plt.tight_layout()
```
