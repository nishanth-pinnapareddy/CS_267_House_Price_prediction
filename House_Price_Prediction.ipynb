## Calculate Percentage of missing values for each varaiable in train data set

```python
for varaible in house_training_data.column_names():
    percent_missing_values = (float(house_training_data[varaible].num_missing())/ len(house_training_data[varaible])) * 100.0
    print ("{0}: \t{1}".format(varaible, percent_missing_values))
```

## Calculate the percentage of missing values in train data set

```python
num_of_missing_values = 0.0
total_no_of_values = 0.0
for varaible in house_training_data.column_names():
    num_of_missing_values += house_training_data[varaible].num_missing()
    total_no_of_values += len(house_training_data[varaible])
percent_missing_values = num_of_missing_values / total_no_of_values * 100.0
print "Percentage of Missing Values in Train Dataset: \t {0}".format(percent_missing_values)
```

## Calculate percentage of missing values for each variable in test data set
